```python
# Define the Leslie matrix and initial population
A = np.array([[0, 1, 5], [0.3, 0, 0], [0, 0.5, 0]])
n_t = np.array([500, 500, 0])


years = 150
pop_by_age = [n_t.copy()]  # Store population for each year

for _ in range(years):
    n_t = np.dot(A, n_t)
    pop_by_age.append(n_t.copy())

# Convert to numpy array for easier handling
pop_by_age = np.array(pop_by_age)

# Print total population for each year
total_population = pop_by_age.sum(axis=1)
print(total_population)
```

```python
import matplotlib.pyplot as plt

# Create a list of years for the x-axis
years_list = range(len(pop_by_age))

# Plot each age class
plt.plot(years_list, pop_by_age[:, 0], label='Age 1')
plt.plot(years_list, pop_by_age[:, 1], label='Age 2')
plt.plot(years_list, pop_by_age[:, 2], label='Age 3')

# Plot total population
plt.plot(years_list, total_population, label='Total Population', linestyle='--', color='black')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Age-Structured Beetle Population Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Define the Leslie matrix and initial population
# Adjust Leslie matrix to reflect 5% removal by traps
A = np.array([[0, 0.95, 4.75], [0.285, 0, 0], [0, 0.475, 0]])
n_t = np.array([500, 500, 0])


years = 150
pop_by_age = [n_t.copy()]  # Store population for each year

for _ in range(years):
    n_t = np.dot(A, n_t)
    pop_by_age.append(n_t.copy())

# Convert to numpy array for easier handling
pop_by_age = np.array(pop_by_age)

# Print total population for each year
total_population_withtrap = pop_by_age.sum(axis=1)
print(total_population_withtrap)
```

```python
# Create a list of years for the x-axis
years_list = range(len(pop_by_age))

# Plot each age class
plt.plot(years_list, pop_by_age[:, 0], label='Age 1')
plt.plot(years_list, pop_by_age[:, 1], label='Age 2')
plt.plot(years_list, pop_by_age[:, 2], label='Age 3')

# Plot total population
plt.plot(years_list, total_population_withtrap, label='Total Population', linestyle='--', color='black')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Age-Structured Beetle Population Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. What difference do you see between the two plots?
    Without traps, the beetle populatin increases rapdily exponentially over time (we didnt include a populatin cap). With traps, the populatin begins declining towards extinction.
2. How much of an effect did seeting the traps have on the beetle population over 150 years?
    The traps do results in the beetles dying out, with more rapid decline starting around 20 years and at 150 when teh simulatin ends the total population with traps is 8 beetles.

```python
# Find the first year when total population with traps reaches 0 or below
year_extinct = None
for i, pop in enumerate(total_population_withtrap):
    if pop <= 1:  # because you cant have half a beetle
        year_extinct = i
        break

if year_extinct is not None:
    print(f"Total population with traps reaches zero at year {year_extinct}")
else:
    print("Population never reaches zero in the simulated years.")
```
