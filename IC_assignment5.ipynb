Creating an SIR model. Test over 30 days.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create an SIR model and simulate an epidemic over 30 days
def sir_model(S0, I0, R0, beta, d, time):
    S_values = [S0]
    I_values = [I0]
    R_values = [R0]

    current_S = S0
    current_I = I0
    current_R = R0

    for day in range(time):
        new_infected = beta * current_S * current_I
        new_recovered = d * current_I

        next_S = current_S - new_infected
        next_I = current_I + new_infected - new_recovered
        next_R = current_R + new_recovered

        S_values.append(next_S)
        I_values.append(next_I)
        R_values.append(next_R)

        current_S, current_I, current_R = next_S, next_I, next_R

    return S_values, I_values, R_values
```
