Plotting normal distributino data

```python
import numpy as np

# Generate normally distributed data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create range of x values for PDF
x = np.linspace(min(data), max(data), 1000)

# Compute theoretical PDF for the same mean and std as the data
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

# Plot histogram of data normalized as density
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Data histogram')

# Plot theoretical PDF
plt.plot(x, pdf, 'r-', lw=2, label='Normal PDF')

# Add labels and legend
plt.title('Normal Distribution and Probability Density')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Suppose Î» (mean rate) = 4 turtles per night
lam = 4

# Generate 1000 random nights of turtle counts
np.random.seed(42)
turtle_counts = np.random.poisson(lam=lam, size=1000)
```

```python
# Define the range of possible turtle counts
x = np.arange(0, np.max(turtle_counts) + 1)

# Theoretical Poisson probability mass function
pmf = poisson.pmf(x, mu=lam)

# Plot histogram normalized to form a probability density
plt.hist(turtle_counts, bins=range(0, np.max(turtle_counts) + 2),
         density=True, alpha=0.6, color='lightgreen', edgecolor='black', label='Observed data')

# Overlay the theoretical Poisson PMF
plt.plot(x, pmf, 'ro-', lw=2, label='Poisson PMF (Î»=4)')

# Labels and title
plt.title('ð¢ Turtle Arrivals per Night (Poisson Model)')
plt.xlabel('Number of Turtles Arriving in a Night')
plt.ylabel('Probability Density (PMF)')
plt.legend()
plt.show()
```
