```python
def N_exp(N_0, p, time):
    # exponential growth function that returns a list of N values over time
    # save N-values into a list
    N_values = [N_0]
    # Start with the initial value
    current_N = N_0
    #set current value to initial value
    for _ in range(time):
        # Iterate for each time step
        current_N = current_N + p * current_N
        # Apply growth
        current_N = round(current_N, 3)
         # round to 3 decimal places
        N_values.append(current_N)
    return N_values
```

```python
result = N_exp(0.01, 0.4, 35)
print(result)
# plot the results with the days on the x axis and the population size on the y axis
import matplotlib.pyplot as plt

days = list(range(36)) # 0 to 36 days
plt.plot(days, result, marker='o')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.title('Exponential Growth of Population Over Time')
plt.grid()
plt.show()
```

```python
#Change N_Exp to N-exp2 and have 3 conditions based on N_0
#If N_0 < 500, p = 0.4
#If N_0 >= 500 and N_0 < 1000, p = 0.2
#If N_0 >= 1000, p = 0

def N_exp2(N_0, time):
    # exponential growth function that returns a list of N values over time
    # save N-values into a list
    N_values = [N_0]
    # Start with the initial value
    current_N = N_0
    #set current value to initial value
    for _ in range(time):
        # Determine p based on current_N
        if current_N < 500:
            p = 0.4
        elif 500 <= current_N < 1000:
            p = 0.2
        else:
            p = 0
        # Apply growth
        current_N = current_N + p * current_N
        current_N = round(current_N, 3)
         # round to 3 decimal places
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with N_0 = 0.01 and time = 35
result = N_exp2(1, 35)
print(result)
```

```python
# plot the results with the days on the x axis and the population size on the y axis

days = list(range(36)) # 0 to 36 days
plt.plot(days, result, marker='o')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.title('Exponential Growth of Population Over Time')
plt.grid()
plt.show()
```
