```python
def N_exp(N_0, p, time):
    # exponential growth function that returns a list of N values over time
    # save N-values into a list
    N_values = [N_0]
    # Start with the initial value
    current_N = N_0
    #set current value to initial value
    for _ in range(time):
        # Iterate for each time step
        current_N = current_N + p * current_N
        # Apply growth
        current_N = round(current_N, 3)
         # round to 3 decimal places
        N_values.append(current_N)
    return N_values
```

```python
#change N_exp to a logistic growth function
def N_logistic(N_0, p, K, epsilon=0.001):
    # logistic growth function that returns a list of N values
    # save N-values into a list
    N_values = [N_0]
    # Start with the initial value
    current_N = N_0
    #set current value to initial value
    while True:
        # Iterate for each time step
        current_N = current_N + p * current_N * (1 - (current_N / K))
        # Apply growth
        current_N = round(current_N, 3)
         # round to 3 decimal places
        N_values.append(current_N)
        if abs(current_N - K) < epsilon:
            break   # Stop if carrying capacity is reached  
    return N_values
```

```python
#Run the function with N_0 = 12, p = 0.6, K = 400
result = N_logistic(12, 0.6, 400)
print(result)
```

```python
import matplotlib.pyplot as plt

# Create a list of time steps to match each population value
time_steps = list(range(len(result)))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time_steps, result, marker='o', linestyle='-', color='green', label='Population over time')
plt.title('Logistic Growth Curve')
plt.xlabel('Time Step')
plt.ylabel('Population (N)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#change N_exp to a logistic growth function
def N_exp3(N_0, K, epsilon=0.001, max_steps=1000):
    # logistic growth function that returns a list of N values
    # save N-values into a list
    N_values = [N_0]
    # Start with the initial value
    current_N = N_0
    #set current value to initial value
    for _ in range(max_steps):
        # Iterate for each time step
        current_N = current_N + current_N * (1 - (current_N / K)) - 0.4 * current_N
        # Apply growth
        current_N = round(current_N, 3)
         # round to 3 decimal places
        N_values.append(current_N)
        if abs(current_N - K) < epsilon:
            break   # Stop if carrying capacity is reached  
    return N_values
```

```python
#Run the function with N_0 = 12, K = 400
result = N_exp3(12, 400)
print(result)
```

```python
# Create a list of time steps to match each population value
time_steps = list(range(len(result)))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time_steps, result, marker='o', linestyle='-', color='green', label='Population over time')
plt.title('Logistic Growth Curve')
plt.xlabel('Time Step')
plt.ylabel('Population (N)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
