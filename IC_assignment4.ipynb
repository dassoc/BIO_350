algorithm to caluclate 2 seperate species to visualize interspecific competition
building on last model of one species, adding a threatend species

```python
def NandM_population(N_0, M_0, time, K_N, K_M,):
    # growth model that returns a list of N values
    # save N-values into a list
    N_values = [N_0]
    M_values = [M_0]
    # Start with the initial value
    current_N = N_0
    current_M = M_0
    #set current value to initial value for N and M
    for _ in range(time):
        # Iterate for each time step
        current_N = current_N + current_N * (1 - (current_N / K_N)-(current_M / K_M)) - 0.4 * current_N
        current_M = current_M + current_M * (1 - (current_M / K_M)-(current_N / K_N)) - 0.6 * current_M
        # Apply growth for N and M species
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
         # round to 3 decimal places
        N_values.append(current_N)  
        M_values.append(current_M)
        # Append the new value to the list
    return N_values, M_values
```

```python
#Run the function with N_0 = 12, K_N = 400, M_0 = 5, K_M = 1600, time = 50
Npop, Mpop = NandM_population(12, 5, 50, 400, 1600)
#do Npop, Mpop to see the output lists easier
print(Npop, Mpop)
```

```python

import matplotlib.pyplot as plt

# Create a list of time steps to match each population value
time_steps = list(range(len(Npop)))

# Plotting
plt.figure(figsize=(10, 6))
# Create a figure with specified size
plt.plot(time_steps, Npop, marker='o', linestyle='-', color='green', label='Species N')
#add markers and lines for N
plt.plot(time_steps, Mpop, marker='s', linestyle='-', color='blue', label='Species M')
#add markers and lines for M
plt.title('Lotka-Volterra Competition: Population Over Time')
#add title
plt.xlabel('Time Step')
#label x axis
plt.ylabel('Population')
#label y axis
plt.grid(True)
#add a grid
plt.legend()
#add a legend for M and N
plt.tight_layout()
#adjust layout to prevent clipping
plt.show()
# Show the plot
```
